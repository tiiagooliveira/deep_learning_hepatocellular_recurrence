```python
print('Number of records:',tcga_raw.shape[0])
print('Number of features:',tcga_raw.shape[1])

print('Data types')
print(tcga_raw.dtypes)
```

```python
binary=[0,1]

for column in tcga_raw.columns[2:]:
    print(column)
    print('NA values:',tcga_raw.shape[0]-tcga_raw[column].count())
    #print('NA values:',tcga_raw.shape[0]-tcga_raw[column].count()+tcga_raw[column].eq('nan').sum())
    if (tcga_raw[column].dtypes=='int64' and len((set(tcga_raw[column].unique())-set(binary)))!=0) or (tcga_raw[column].dtypes=='float'):
        print('Max:',tcga_raw[column].max())
        print('Min:',tcga_raw[column].min())
        print('Mean:',tcga_raw[column].mean())
        print('Var:',tcga_raw[column].var())
        print('Std:',tcga_raw[column].std())
              
    elif tcga_raw[column].dtypes=='int64' and len((set(tcga_raw[column].unique())-set(binary)))==0:
        print('Number of positives:',tcga_raw[column].tolist().count(1))
        print('Number of negatives:',tcga_raw[column].tolist().count(0))
    else: 
        values=list(tcga_raw[column].unique())
        print('Mode:',tcga_raw[column].mode()[0])
        print('Values:')
        print(*values,sep=', ')
    print('---------------------')
    
```
